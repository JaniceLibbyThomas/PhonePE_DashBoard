#converting Aggregated TRANSACTION to DF

**TRANSACTION INDIA:**
TIME : 8S

```python
import pandas as pd
import requests

#list of year folder that we have to parse into
year = ['2018','2019','2020','2021','2022']
c = 0
for Y in year :
  for json_file in range (1,5):
    url = f"https://raw.githubusercontent.com/PhonePe/pulse/master/data/aggregated/transaction/country/india/{Y}/{json_file}.json"
    r = requests.get(url)

    data = r.json()
    entries = data["data"]["transactionData"]


    value_type = []
    value_count = []
    value_amount = []
    value_name =[]
    value_from=[]
    value_to = []
    value_year = []
    value_Quater = []
    for i in range (0, len(entries)):
      value_from.append(data['data']['from'])
      value_to.append(data['data']['to'])
      value_name.append(data['data']['transactionData'][i]['name'])
      value_type.append(entries[i]['paymentInstruments'][0]['type'])
      value_count.append(entries[i]['paymentInstruments'][0]['count'])
      value_amount.append(entries[i]['paymentInstruments'][0]['amount'])
      value_year.append(Y)
      value_Quater.append('Q'+str(json_file))
    #creating new dataframe and appending datas.

    temp_df = pd.DataFrame(
    {'Data From': value_from,
     'Data To': value_to,
     'Name': value_name,
     'Type': value_type,
     'Count': value_count,
     'Amount': value_amount,
     'Year': value_year,
     'Quater': value_Quater
     
    })

    if(c ==0):
      Transaction_India = pd.DataFrame(temp_df)
      c += 1
    else:
      Transaction_India =  Transaction_India.append([temp_df])


Transaction_India.reset_index(inplace = True, drop = True)
```

```python
Transaction_India.to_csv('MAP_TRANS_AGG_INDIA.csv')
```
